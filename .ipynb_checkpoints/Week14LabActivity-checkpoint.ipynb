```python
data = pd.read_csv(r"/Users/avedon/Documents/GitHub/E11-Lab-Besties/419test2.csv")
data
# after row 31 the gps starts working
data_gps = data[32:80]
```

```python
plt.plot(data['Epoch time'], data["Air Quality PM 2.5 (ug/m3)"])
plt.ylabel("Air Quality (ug/m3)")
plt.xlabel("Epoch Time")
plt.show()

plt.plot(data_gps["Latitude"], data_gps["Air Quality PM 2.5 (ug/m3)"])
plt.ylabel("Air Quality (ug/m3)")
plt.xlabel("Latitude")
plt.show()

plt.plot(data_gps['Longitude'], data_gps["Air Quality PM 2.5 (ug/m3)"])
plt.ylabel("Air Quality (ug/m3)")
plt.xlabel("Longitude")
plt.show()

plt.plot(data['Epoch time'], data["Temperature (C)"])
plt.ylabel("Temperature (C)")
plt.xlabel("Epoch Time")
plt.show()

plt.plot(data_gps["Latitude"], data_gps["Temperature (C)"])
plt.ylabel("Temperature (C)")
plt.xlabel("Latitude")
plt.show()

plt.plot(data_gps['Longitude'], data_gps["Temperature (C)"])
plt.ylabel("Temperature (C)")
plt.xlabel("Longitude")
plt.show()
```

```python
plt.hist(data["Air Quality PM 2.5 (ug/m3)"], bins=50)
plt.xlabel("Air Quality (ug/m3)")
plt.ylabel("Frequency")
plt.show()

plt.hist(data_gps["Air Quality PM 2.5 (ug/m3)"], bins=50)
plt.xlabel("Air Quality (ug/m3)")
plt.ylabel("Frequency")
plt.show()

plt.hist(data_gps["Air Quality PM 2.5 (ug/m3)"], bins=50)
plt.xlabel("Air Quality (ug/m3)")
plt.ylabel("Frequency")
plt.show()

plt.hist(data["Temperature (C)"], bins=50)
plt.xlabel("Temperature (C)")
plt.ylabel("Frequency")
plt.show()

plt.hist(data_gps["Temperature (C)"], bins=50)
plt.xlabel("Temperature (C)")
plt.ylabel("Frequency")
plt.show()

plt.hist(data_gps["Temperature (C)"], bins=50)
plt.xlabel("Temperature (C)")
plt.ylabel("Frequency")
plt.show()
```

```python
#AQ PM 2.5 (ug/m3)
aq_mean = np.mean(data["Air Quality PM 2.5 (ug/m3)"])
aq_std = np.std(data["Air Quality PM 2.5 (ug/m3)"])
print("Air Quality PM 2.5 (ug/m3) mean:", aq_mean)
print("Air Quality PM 2.5 (ug/m3) standard deviation:", aq_std)

#AQ PM 2.5 (ug/m3) with respect to latitude
aq_lat_mean = np.mean(data_gps["Air Quality PM 2.5 (ug/m3)"])
aq_lat_std = np.std(data_gps["Air Quality PM 2.5 (ug/m3)"])
print("Air Quality PM 2.5 (ug/m3) with respect to latitude mean:", aq_lat_mean)
print("Air Quality PM 2.5 (ug/m3) with respect to latitude standard deviation:", aq_lat_std)

#AQ PM 2.5 (ug/m3) with respect to longitude
aq_long_mean = np.mean(data_gps["Air Quality PM 2.5 (ug/m3)"])
aq_long_std = np.std(data_gps["Air Quality PM 2.5 (ug/m3)"])
print("Air Quality PM 2.5 (ug/m3) with respect to longitude mean:", aq_long_mean)
print("Air Quality PM 2.5 (ug/m3) with respect to longitude standard deviation:", aq_long_std)

# Temp (C)
temp_mean = np.mean(data["Temperature (C)"])
temp_std = np.std(data["Temperature (C)"])
print("Temperature (C) mean:", temp_mean)
print("Temperature (C) standard deviation:", temp_std)

# Temp (C) with respect to latitude
temp_lat_mean = np.mean(data_gps["Temperature (C)"])
temp_lat_std = np.std(data_gps["Temperature (C)"])
print("Temperature (C) with respect to latitude mean:", temp_lat_mean)
print("Temperature (C) with respect to latitude standard deviation:", temp_lat_std)

# Temp(C) with respect to longitude
temp_long_mean = np.mean(data_gps["Temperature (C)"])
temp_long_std = np.std(data_gps["Temperature (C)"])
print("Temperature (C) with respect to longitude mean:", temp_long_mean)
print("Temperature (C) with respect to longitude standard deviation:", temp_long_std)
```
